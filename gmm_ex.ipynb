```python
import numpy as np
from matplotlib import pylab as plt
```

```python
from scipy.cluster.vq import kmeans2
from gauss import Gauss
from  kmeans import Kmeans
from gmm import Gmm
```

```python
m1 = np.array([0.,0.]); m2 = np.array([-2.5,-2.5]); m3 = np.array([2.5,2.5])
c1 = np.array([[2.,1.],[1.,2.]]); c2 = np.array([[.1,0.],[0.,5.0]]); c3 = np.array([[2.,0],[0,.1]])

g1 = Gauss(m1,c1)
g2 = Gauss(m2,c2)
g3 = Gauss(m3,c3)

dat1 = g1.sample(100)
dat2 = g2.sample(100)
dat3 = g3.sample(100)

plt.hold('on')
plt.plot(dat1[0,:],dat1[1,:],'.r')
plt.plot(dat2[0,:],dat2[1,:],'.b')
plt.plot(dat3[0,:],dat3[1,:],'.g')
plt.title('The three classes of the training set.')


data = np.hstack((dat1,dat2))
data = np.hstack((data,dat3))
plt.figure()
plt.plot(data[0,:],data[1,:],'.k')
plt.title('The data without class indicators.')
```

```python
g = Gmm(data,k=3)
```

```python
prob = g.soft_assign(data)
k,n = prob.shape
for i in range(n):
    plt.plot(data[0,i],data[1,i],'.',color=prob[:,i])
plt.title('The soft class membership assigned by GMM.')
```
